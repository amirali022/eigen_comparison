## Comparison of Eigen Values of a Multivariate Normal Distribution Before and After Applying Noise

### import statements

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Constants

```python
MEAN = [ 3, 0, 5]
COVARIANCE = [
	[ 2, 0, -1],
	[ 0, 1, 0],
	[ -1, 0, 2]
]
```

### Generate Samples From Multivariate Normal Distribution Given Mean Vector and Covariance Matrix

```python
samples = np.random.default_rng().multivariate_normal( MEAN, COVARIANCE, 100)

x, y, z = samples.T
```

### Scatter Plot of Samples

```python
fig = plt.figure( figsize=( 10, 10))
ax = fig.add_subplot( projection="3d")

ax.scatter( x, y, z)

plt.show()
```

### Calculate Mean of Samples

```python
mean = samples.mean( axis=0)

print( f"Mean Vector:\n{ mean}")
```

### Calculate Covariance of Samples

```python
covariance = np.cov( samples, rowvar=False)

print( f"Covariance Matrix:\n{ covariance}")
```

### Calculate Eigen Values and Eigen Vectors of Covariance Matrix of Samples

```python
eigen_values, eigen_vectors = np.linalg.eig( covariance)

print( f"Eigen Values:\n{ eigen_values}")
print( f"Eigen Vectors:\n{ eigen_vectors}")
```

### Generate Noise from Normal Distribution N(0,1)

```python
noise = np.random.normal( 0, 1, ( 100, 3))
```

### Add Noise to Samples

```python
noisy_samples = samples + noise
```

### Calculate Mean of Noisy Samples

```python
noisy_mean = noisy_samples.mean( axis=0)

print( f"Noisy Mean:\n{ noisy_mean}")
```

### Calculate Covarince of Noisy Samples

```python
noisy_covariance = np.cov( noisy_samples, rowvar=False)

print( f"Noisy Covariance:\n { noisy_covariance}")
```

### Calculate Eigen Values and Eigen Vectors of Covariance Matrix of Noisy Samples

```python
noisy_eigen_values, noisy_eigen_vectors = np.linalg.eig( noisy_covariance)

print( f"Noisy Eigen Values:\n{ noisy_eigen_values}")
print( f"Noisy Eigen Vectors:\n{ noisy_eigen_vectors}")
```

### Calculate Difference of Eigen Values of Real Samples and Noisy Samples

```python
eigen_values_differences = eigen_values - noisy_eigen_values

eigen_values_differences
```
